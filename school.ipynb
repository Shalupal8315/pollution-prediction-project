```python
import pandas as pd

school = pd.read_csv("school.csv", encoding="latin1")

print(school.columns.tolist())

print(school.head())
```

```python
import re

school["Pincode"] = school["Address"].astype(str).str.extract(r'(\d{6})')

print(
    school[
        ["School Name", "Address", "Pincode"]
    ].head()
)
```

```python
print("Total Schools:", len(school))

print(
    "Schools with Pincode:",
    school["Pincode"].notna().sum()
)

print(
    "Schools without Pincode:",
    school["Pincode"].isna().sum()
)
```

```python
school_count = (
    school.groupby("Pincode")
    .size()
    .reset_index(name="school_count")
)

print(school_count.head())

print("Total Unique Pincodes:",
      len(school_count))
```

```python
stations = pd.read_csv("station_summary.csv")

stations["pincode"] = stations["pincode"].astype(str)
school_count["Pincode"] = school_count["Pincode"].astype(str)

stations = stations.merge(
    school_count,
    left_on="pincode",
    right_on="Pincode",
    how="left"
)

stations["school_count"] = (
    stations["school_count"]
    .fillna(0)
)

print(
    stations[
        [
            "station",
            "pincode",
            "school_count"
        ]
    ].head()
)
```

```python
import pandas as pd

station_summary = pd.read_csv("station_summary.csv")

print(station_summary.head())
```

```python
def risk_level(aqi):

    if aqi >= 290:
        return "Critical"

    elif aqi >= 280:
        return "High"

    elif aqi >= 260:
        return "Moderate"

    else:
        return "Low"
```

```python
station_summary["risk_level"] = (
    station_summary["aqi"]
    .apply(risk_level)
)

station_summary.to_csv(
    "station_summary.csv",
    index=False
)

print(
    station_summary[
        ["station", "aqi", "risk_level"]
    ]
)
```
