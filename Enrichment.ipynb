```python
import pandas as pd
import requests
import time

stations = pd.read_csv("station_master.csv")

pincodes = []

for _, row in stations.iterrows():

    lat = row["latitude"]
    lon = row["longitude"]

    url = f"https://nominatim.openstreetmap.org/reverse?format=jsonv2&lat={lat}&lon={lon}"

    headers = {
        "User-Agent": "DelhiNCRAQIProject"
    }

    try:

        r = requests.get(
            url,
            headers=headers
        )

        data = r.json()

        pincode = data.get(
            "address",
            {}
        ).get(
            "postcode",
            ""
        )

        pincodes.append(pincode)

    except:

        pincodes.append("")

    time.sleep(1)

stations["pincode"] = pincodes

stations.to_csv(
    "station_with_pincode.csv",
    index=False
)
```
